```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Literal
```

```python
class QuadEqnState(TypedDict):
    a: float
    b: float
    c: float
    discriminant: float
    equation: str
    result: str
```

```python
def show_equation(state: QuadEqnState) -> QuadEqnState:
    a = state['a']
    b = state['b']  
    c = state['c']
    return {'equation': f"{a}x^2 + {b}x + {c} = 0"}

def calc_discriminant(state: QuadEqnState) -> QuadEqnState:
    a = state['a']
    b = state['b']  
    c = state['c']
    return {'discriminant': b**2 - 4*a*c}

def real_roots(state: QuadEqnState) -> QuadEqnState:
    a = state['a']
    b = state['b']  
    c = state['c']
    d = state['discriminant']
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    return {'result': f"Roots are real: {root1} and {root2}"}

def repeated_roots(state: QuadEqnState) -> QuadEqnState:
    a = state['a']
    b = state['b']  
    root = -b / (2*a)
    return {'result': f"Roots are repeated: {root}"}

def complex_roots(state: QuadEqnState) -> QuadEqnState:
    result = f"No real roots"
    return {'result': result}

def check_condition(state: QuadEqnState) -> Literal['real_roots', 'repeated_roots', 'complex_roots']: # output is name of next node (function) based on condition check
    d = state['discriminant']
    if d > 0:
        return 'real_roots'
    elif d == 0:
        return 'repeated_roots'
    else:
        return 'complex_roots'
```

```python
graph = StateGraph(QuadEqnState)

graph.add_node('show_equation', show_equation)
graph.add_node('calc_discriminant', calc_discriminant)
graph.add_node('real_roots', real_roots)
graph.add_node('repeated_roots', repeated_roots)
graph.add_node('distinct_roots', complex_roots)

graph.add_edge(START, 'show_equation')
graph.add_edge('show_equation', 'calc_discriminant')

graph.add_conditional_edge('calc_discriminant',check_condition)
graph.add_edge('real_roots', END)
graph.add_edge('repeated_roots', END)   
graph.add_edge('complex_roots', END)

workflow = graph.compile()
```
